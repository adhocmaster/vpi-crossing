```python
import numpy as np
import matplotlib.pyplot as plt
```

### Normal Distribution of Pedestrian Gap Acceptance.

```python
mu = 4.0
sigma = 2.5
s = np.random.normal(mu, sigma, 1000)
count, bins, ignored = plt.hist(s, 30, density = 1)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
         np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.rcParams['figure.facecolor'] = 'white'
plt.show()
```
